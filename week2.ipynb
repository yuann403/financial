1.證券交易所

```python
import requests
import pandas as pd
import datetime as dt
```

```python
date = dt.datetime.now().strftime("%Y%m%d")
stock = "2330"
print("今天日期:",date)
print("股票代號:",stock)

url = f"https://www.twse.com.tw/rwd/zh/afterTrading/STOCK_DAY?date={date}&stockNo={stock}"

response = requests.get(url)
data = response.json()

df = pd.DataFrame(data = data["data"], columns = data["fields"])
df
```

2.Yahoo奇摩

```python
from bs4 import BeautifulSoup as bs
```

```python
stock = "2330"
url = f"https://tw.stock.yahoo.com/quote/{stock}.TW"

response = requests.get(url)
soup = bs(response.text, "html.parser")

time = soup.find("section", {"id": "qsp-overview-realtime-info"}).find("time")
price = soup.find_all('li', class_='price-detail-item')

data = {}
for item in price:
  label = item.find('span').text
  value = item.find_all('span')[1].text
  data[label] = value

df = pd.DataFrame([data])
df.index = [time['datatime']]
print("股票代號:",stock)
df
```
